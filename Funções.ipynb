```python
def calculo_do_total(unidade, quantidade):
  total = unidade * quantidade
  return total
```

```python
unidade = float(input("Qual é o valor da unidade"))
quantidade = float(input('Qual é a quantidade de produtos'))
valor = calculo_do_total(unidade, quantidade)
print(f'O valor total é R${valor}')
```

```python
def calculo_imc(peso, altura):
  imc = peso / (altura ** 2)
  return imc
```

```python
peso = float(input('Qual é peso: '))
altura = float(input('Qual é a sua altura: '))

resultado = calculo_imc(peso, altura)
print(f'Seu IMC é igual {resultado}')
```

```python
def calculo_média(a, b, c):
  soma = a + b + c
  média = soma / 3
  return média, soma
```

```python
 n1 = float(input( 'Qual foi sua primeira nota?: ') )
 n2 = float(input( 'Qual foi sua segunda nota?: ') )
 n3 = float(input( 'Qual foi sua terceira nota?: ') )

resu_media, resu_soma = calculo_média(n1, n2, n3)
print(f'A soma dos valores é de {resu_soma}')
print(f'A média é de {resu_media}')
```

```python
#6 EXE
def calcular_desconto(valor_original, percentual_desconto):
  VF = valor_original * (1 - (percentual_desconto / 100))
  return VF

VO = float(input("Qual é o valor original?: "))
PD = float(input('Qual é a porcentagem de desconto? (sem o sinal de %): '))

num = calcular_desconto(VO, PD)
print(f'O valor final já com o desconto é de {num}')
```

```python
#7 EXE
def max_min(lista):
    Max = max(lista)
    Min = min(lista)
    return Max, Min

num_usu = input('Me fale uma lista de números separados por espaços: ')
lista_usu = [float(num) for num in num_usu.split()]

resultado = max_min(lista_usu)
print(f'Os números máximo e mínimo são: {resultado}')
```
