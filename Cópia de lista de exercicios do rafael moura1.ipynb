```python
#8 Faça um programa que mostre os n termos da Série a seguir e realize o cálculo Solicite ao usuário a quantidade de repetições
n = int(input("Digite a quantidade de termos da série: "))

soma = 0
denominador = 1

print("Série:")

for numerador in range(1, n+1):
    termo = numerador / denominador
    print(f"{numerador}/{denominador}", end=" ")
    if numerador != n:
        print("+", end=" ")
    soma += termo
    denominador += 2

print(f"\nSoma da série: {soma}")
```

```python
#6
palavras_negativas = [
    'triste', 'deprimido', 'ansioso', 'desapontado', 'frustrado',
    'preocupado', 'inseguro', 'solitário', 'desanimado', 'estressado'
]

texto = input("Digite um texto para análise: ").lower()
palavras_texto = texto.split()
negativas_encontradas = []
for palavra in palavras_texto:
    if palavra in palavras_negativas:
        negativas_encontradas.append(palavra)
if negativas_encontradas:
    print("Palavras negativas encontradas no texto:", negativas_encontradas)
else:
    print("Nenhuma palavra negativa encontrada no texto.")
```
